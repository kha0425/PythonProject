## 2 막데 그래프 - bar() 세로

```python
# 운동 시작전과 운동 한달 후의 윗몸 일으키기 횟수 그래프
member_ids =['id_01','id_02','id_03','id_04'] # 회원 ID
before_ex = [27,35,40,33] # 운동 시작 전
after_ex = [30,38,42,37] # 운동 한 달 후

# numpy를 이용해서 배열 생성
index = np.arange(len(member_ids))
index
```

```python
# 운동 시작 전
plt.bar(index, before_ex) # bar(x,y) x = index, height = before_ex
plt.show()
```

```python
# x축 -> tick_label 옵션 사용 -> id로 지정정
plt.bar(index, before_ex, tick_label = member_ids)
plt.show()
```

```python
# 막대그래프 색깔 -> color옵션
colors = ['r','g','b','m']
plt.bar(index, before_ex, color = colors, tick_label = member_ids)
plt.show()
```

```python
# 막대그래프의 폭을 변경 -> width옵션
plt.bar(index, before_ex, color = colors, tick_label = member_ids, width = 0.6)
plt.show()
```

## 3 가로 막대 그래프 - barh()

```python
plt.barh(index, before_ex, color = colors, tick_label = member_ids) # width옵션 사용 못함
plt.show()
```

### 막대 그래프 실습 - 운동 시작전과 한 달 후의 비교 막대 그래프

```python
# bar()의 width옵션 두 개의 막대그래프가 들어갈 수 있도록 여유있게 너비 설정 ---> 0.4
bar_width = 0.4
```

```python
# bar() 함수의 align옵션 --> 그래프를 한쪽으로 치우치게 한다. -> edge
plt.bar(index, before_ex, color = 'c', align = 'edge', width = bar_width, label='before')
plt.show()
```

```python
# 운동 한 달 후 그래프는 겹치지 않게 x 좌표에서 막대 그래프 시작 위치를 두께(너비)만큼 오른쪽으로 0.4 이동
plt.bar(index+bar_width ,after_ex, color = 'm', align='edge',width=bar_width,label='after')
plt.show()
```

```python
bar_width = 0.4

plt.bar(index,before_ex,color='c',align='edge',width=bar_width,label='before')

plt.bar(index+bar_width,after_ex,color='m',align='edge',width=bar_width,label='after')

# xticks(tick 위치, tick_label)
plt.xticks(index + bar_width, member_ids)
plt.legend() # 범례명이 label이 사용됨
plt.xlabel('회원  ID')
plt.ylabel('윗몸 일으키기 횟수')
plt.title('운동 시작 전과 한 달 후의 근지구력(근력) 변화 비교')
plt.show()
```

```python
# x - > 경도(lon), y -> 위도(lat)
# 투명도를 중간으로 설정 -> alpha
plt.scatter(lon,lat,s=size,c=colors,alpha=0.5)
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017년)')

# 마커가 위치한 곳의 도시 이름 표시
for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name) # 위도 경도에 맞게 도시 이름 출력
plt.show() 
```
