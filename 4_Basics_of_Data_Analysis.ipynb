## 3 파생변수 만들기

### 변수 조합해 파생변수 만들기

```python
df = pd.DataFrame({'var1' :[4,3,8],
                   'var2' : [2,6,1]})
df
```

```python
# var_sum 파생변수 만들기
df['var_sum'] = df['var1'] + df['var2']
df
```

```python
# var_mean 파생변수 만들기
df['var_mean'] = (df['var1'] + df['var2']) / 2
df
```

### mpg 통합 연비 변수 만들기

```python
# 통합 연비 변수 만들기
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
# total 합계 행 수로 나누기
mpg_total_avg = sum(mpg['total']) / len(mpg)
mpg_total_avg
```

```python
# 통합 연비 변수 평균
mpg_total_avg = mpg['total'].mean()
mpg_total_avg
```

```python
# mpg 데이터 통계값
mpg.describe(include='all')
```

### 조건문을 활용해 파생변수 만들기

```python
#  1. 기준값 정하기
mpg['total'].describe() # 요약 통계량 출력
```

```python
# 그래프 - 히스토그램 만들기
mpg['total'].plot.hist()
```

```python
# 2. 합격 판정 변수 만들기
# - total 변수가 20을 넘으면 합격, 넘지 못하면 불합격으로 분류
import numpy as np # 배열 연산, 통계칠 계산 등 수치연산할 때 사용하는 패키지
```

```python
# where() : 조건에 따라 서로 다른 값을 부여한다.
mpg['test'] = np.where(mpg['total'] >= 20 ,'pass','fail')
mpg.head()
```

```python
mpg.tail()
```

```python
# 3. 빈도표로 합격 판정 자동차 수 살펴보기
# 빈도표 : 변수의 값이 종류별로 몇 개씩 있는지, 값의 개수를 나타낸 표
mpg['test'].value_counts()
```

```python
# 4. 막대그래프로 빈도 표현하기
count_test = mpg['test'].value_counts()
count_test.plot.bar()
count_test.plot.bar(rot=0) # 축 이름 수평으로 만들기
```

### 중첩 조건문 활용하기

```python
# 1. 연비 등급 변수
# total 기준으로 A,B,C 등급 부여
mpg['grade'] = np.where(mpg['total'] >= 30 , 'A',
                        np.where(mpg['total'] >= 20, 'B','C'))
mpg.head()
```

```python
# 2. 빈도표와 막대 그래프로 연비 등급 살펴보기
# 등급 빈도표
count_grade = mpg['grade'].value_counts()
count_grade
count_grade.plot.bar(rot=0)
```

```python
# 등급 빈도표 알파벳 순으로 정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
count_grade.plot.bar(rot=0)
```

```python
# 내림차순으로 정렬
count_grade = mpg['grade'].value_counts().sort_index(ascending=False)
count_grade
```

```python
# A, B, C, D 등급 변수 만들기
mpg['grade2'] = np.where(mpg['total'] >=30, 'A',
                np.where(mpg['total'] >= 25, 'B',
                np.where(mpg['total'] >=20, 'C', 'D')))
mpg.head()
```

```python
mpg['category'].value_counts()
```

```python
mpg['size'] = np.where((mpg['category'] == 'compact')|
                       (mpg['category'] == 'subcompact')|
                       (mpg['category'] == '2seater'),
                       'small','large')
```
